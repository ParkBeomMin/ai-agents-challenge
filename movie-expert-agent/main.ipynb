```python
import openai

client = openai.OpenAI()

DEFAULT_PROMPT = """
You are a movie expert.

I have the following functions
- `get_popular_movies`: receive no arguments. It returns a list of popular movies.
- `get_movie_details`: receive the movie id as an argument (i.e `get_movie_details(movie_id)`). It returns the details of the movie.
- `get_movie_credits`: receive the movie id as an argument (i.e `get_movie_credits(movie_id)`). It returns the credits of the movie.

Please answer with the name of the function that you would like to me.

Please say nothing else. just the name of the function with the arguments.

Answer the following question:
"""
```

```python
PROMPT = DEFAULT_PROMPT + """
지금 인기 있는 영화가 무엇인지 알려줘.
"""

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": PROMPT}],
)

message = response.choices[0].message.content

message
```
