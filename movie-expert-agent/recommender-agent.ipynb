```python
import openai
import os

client = openai.OpenAI()

DEFAULT_PROMPT = f"""
You are a movie expert and a recommender.

I have the following functions
- `get_popular_movies`: receive no arguments. It returns a list of popular movies. you can get the list of popular movies from {os.getenv('MOVIE_API_BASE_URL')}/movies
- `get_movie_details`: receive the movie id as an argument (i.e `get_movie_details(movie_id)`). It returns the details of the movie. you can get the details of the movie from {os.getenv('MOVIE_API_BASE_URL')}/movies/movie_id
- `get_movie_credits`: receive the movie id as an argument (i.e `get_movie_credits(movie_id)`). It returns the credits of the movie. you can get the credits of the movie from {os.getenv('MOVIE_API_BASE_URL')}/movies/movie_id/credits
"""

messages = []

def ai_call(prompt):
    messages.append({"role": "user", "content": prompt})
    print(f"User: {prompt}")
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
    )
    messages.append({"role": "assistant", "content": response.choices[0].message.content})
    print(f"Assistant: {response.choices[0].message.content}")
```

```python
while True:
    message = input("Enter a message: ")
    if message == "exit" or message == "quit" or message == "bye" or message == "q":
        break
    ai_call(message)
```
